## **Problem 1: Social Media – Post Engagement Analyzer**
**Problem Statement**
- Create a function to analyze engagement on a social media post.

**Rules**
- Calculate total likes
- If Total Likes ≥ 1000 → Viral Post
- Otherwise → Normal Engagement

**Real-Time Use**
- Social media analytics platforms
- Marketing dashboards

**Hint**
- Use function
- Use loop for calculations

**Expected Output Format**
- Total Likes: 1050
- Post Status: Viral Post

```python
def Engagement_Analyzer(daily_likes):

    # Checking if likes data is present
    if not daily_likes:
      return 0, "No engagement status."
    # Adding all the likes if each day to total likes variable
    total_likes = 0
    for likes in daily_likes:
      total_likes += likes
    # Loop If Total Likes ≥ 1000 then Viral Post
    if total_likes >= 1000:
      status = "Viral Post"
      return total_likes, status
    # Loop If Total Likes < 1000 then Normal Engagement
    else:
      status = "Normal Engagement"
      return total_likes, status

daily_likes = [150, 20, 350, 450, 600]
total_likes_result, status_result = Engagement_Analyzer(daily_likes)
print(f"Total Likes: {total_likes_result}")
print(f"Enagagement Status: {status_result}")
```

## **Problem 2: Healthcare – Medicine Stock Alert System**
**Problem Statement**
- Create a function to monitor medicine stock levels.

**Rules**
- Stock < 10 → Low Stock Alert
- Stock ≥ 10 → Stock Sufficient

**Real-Time Use**
- Pharmacy management systems
- Hospital inventory tracking

**Hint**
- Use function
- Use conditional logic

**Expected Output Format**
- Medicine Stock: 6
- Status: Low Stock Alert

```python
def medicine_stock_alert(medicine_stock):

  # checking if stock data is present
  if medicine_stock == 0:
    return 0, "No stock alert."

  # checking if Stock < 10 then Low Stock Alert
  if medicine_stock < 10:
    status = "Low Stock Alert"
    return medicine_stock, status
  # Stock ≥ 10 then Stock Sufficient
  else:
    status = "Stock Sufficient"
    return medicine_stock, status

medicine_stock = 6
medicine_stock_result, status_result = medicine_stock_alert(medicine_stock)
print(f"Medicine Stock: {medicine_stock_result}")
print(f"Status: {status_result}")
```

## **Problem 3: Agriculture – Rainfall Adequacy Checker**
**Problem Statement**
- Create a function that analyzes seasonal rainfall data.

**Rules**
- Calculate average rainfall
- Average ≥ required level → Adequate Rainfall
- Otherwise → Inadequate Rainfall

**Real-Time Use**
- Climate monitoring systems
- Irrigation planning tools

**Hint**
- Use function
- Use loop and math operations

**Expected Output Format**
- Average Rainfall: 72
- Rainfall Status: Adequate Rainfall

```python
def Rainfall_Adequacy_Checker(monthly_rainfall):
  total_rainfall = 0

  # Loop to calculate sum of rainfall data
  for rainfall in monthly_rainfall:
    total_rainfall += rainfall

  # Calculate average rainfall
  average_rainfall = total_rainfall / len(monthly_rainfall)

  # loop to distinguish adequate rainfall
  if average_rainfall >= 100:
    status = "Adequate Rainfall"
  else:
    status = "Inadequate Rainfall"

  return average_rainfall, status

monthly_rainfall = [100, 150, 120, 180, 200, 350]
average_rainfall_result, status_result = Rainfall_Adequacy_Checker(monthly_rainfall)
print(f"Average Rainfall: {average_rainfall_result}")
print(f"Rainfall Status: {status_result}")
```

## **Problem 4: Social Media – Duplicate Account Detection**
**Problem Statement**
- Create a function to detect duplicate usernames in a system.

**Rules**
- If duplicates exist → Flag accounts
- Else → Accounts are unique

**Real-Time Use**
- Platform moderation
- Fraud detection systems

**Hint**
- Use function
- Use set data structure

**Expected Output Format**
- Duplicate Accounts Found: Yes


```python
def detect_duplicates(usernames):

    # Convert list to set (removes duplicates automatically)
    unique_usernames = set(usernames)

    # Compare lengths: if different, duplicates exist
    if len(unique_usernames) < len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# Example usage:
usernames = ["alice", "bob", "charlie", "alice", "david"]
detect_duplicates(usernames)
```

**Real-Time Use**

**Real-Time Use**

## **Problem 9: E-Learning – Student Attendance Eligibility System**
**Problem Statement**
- Create a function to calculate attendance percentage and check exam eligibility.

**Rules**
- Attendance ≥ 75% → Eligible
- Attendance < 75% → Not Eligible

**Real-Time Use**
- Learning management systems
- Internship tracking platforms

**Hint**
- Use function
- Use list and loop

**Expected Output Format**
- Attendance Percentage: 80.0
- Exam Eligibility: Eligible

```python
def check_attendance(attended_classes, total_classes):

    # Calculate attendance percentage
    attendance_percentage = (attended_classes / total_classes) * 100

    # Print the attendance percentage
    print(f"Attendance Percentage: {attendance_percentage:.1f} %")

    # Conditional check for eligibility
    if attendance_percentage >= 75:
      Exam_Eligibility = "Eligible"
      print("Exam Eligibility:", Exam_Eligibility)
    else:
      Exam_Eligibility = "Not Eligible"
      print("Exam Eligibility:", Exam_Eligibility)


# Example usage:
attended = 40
total = 50
check_attendance(attended, total)
```

## **Problem 10: Smart Electricity Bill Analyzer**
**Problem Statement**
- You are building a smart electricity billing system for residential users.
- Create a function that calculates the electricity bill and classifies usage level based on units consumed.

**Billing Rules**
- First 100 units → ₹3 per unit
- Next 100 units (101–200) → ₹5 per unit
- Above 200 units → ₹7 per unit

**Usage Classification**
- Total bill < ₹500 → Low Usage
- ₹500 – ₹1500 → Moderate Usage
- ₹1500 → High Usage

**Real-Time Use**
- Smart meter systems
- Electricity board billing software
- Energy consumption analytics

**Hint**
- Use function to calculate bill
- Use conditional logic for slabs
- Use return values (bill and usage status)


```python
def calculate_bill(units):
    # Calculate bill based on slabs
    if units <= 100:
        total_bill = units * 3
    elif units <= 200:
        total_bill = (100 * 3) + (units - 100) * 5
    else:
        total_bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    # Classify usage based on bill amount
    if total_bill < 500:
        usage_status = "Low Usage"
    elif total_bill <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"

    return total_bill, usage_status


# Example usage:
units_consumed = 250
bill, status = calculate_bill(units_consumed)
print(f"Units Consumed: {units_consumed}")
print(f"Total Bill: ₹{bill}")
print(f"Usage Classification: {status}")
```
