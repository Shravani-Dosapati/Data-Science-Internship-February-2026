## **Problem 1: Mobile Recharge Validation System**
**Problem Statement**
- Create a function to validate a mobile recharge request.

**Rules:**
- Recharge amount must be ≥ ₹50

- **Valid plans:** 199, 299, 399, 599

- User should be allowed to retry if invalid


**Real-Time Use**
- Telecom recharge apps


**Payment gateways**

**Hint**
- Use function for validation
- Use while loop for retry

```python
def mobile_recharge_validation(charge):

  valid_plans = [199, 299, 399, 599]

  while True:
    try:
      # Rule 1: Minimum amount check
      if charge < 50:
        return "Recharge amount must be ≥ ₹50"
      # Rule 2: Valid plan check
      elif charge not in valid_plans:
        return "Choose Valid plans (199, 299, 399, 599)"
      else:
        return f"Recharge Successful for Rs.{charge}."
    except ValueError:
      print("Invalid input. Please enter a valid number.")
```

```python
# Run the function
mobile_recharge_validation(399)
```

## **Problem 2: Inventory Reorder Alert System**
**Problem Statement**
- Given a dictionary of products and stock quantities:
  - Print Reorder Alert if stock < 15


  - Print Stock OK otherwise


**Real-Time Use**
- Warehouse systems


- Retail management


**Hint**
- Use function


- Use for loop on dictionary

```python
def inventory_alert(stock_dict):
    for product, quantity in stock_dict.items():
        if quantity < 15:   # Setting quantity constraint to more than 15
            print(f"{product}: Reorder Alert (Stock = {quantity})")
        else:
            print(f"{product}: Stock OK (Stock = {quantity})")

# Example dictionary of products and stock quantities
inventory = {
    "Apples": 10,
    "Bananas": 25,
    "Oranges": 14,
    "Mangoes": 30
}

# Run the function
inventory_alert(inventory)
```

## **Problem 3: Student Result Processing System**
**Problem Statement**
- Create a function that:
  - Accepts a list of student marks


  - Calculates average marks


  - Prints Pass if average ≥ 50 else Fail


**Real-Time Use**
- Online exam systems


- Evaluation portals


**Hint**
- Use loop to calculate total


- Return result from function


```python
def process_result(marks):
    # Calculate total using a loop
    total = 0
    for mark in marks:
        total += mark

    # Calculate average
    average = total / len(marks)

    # Print result based on average
    if average >= 50:
        return "Pass"
    else:
        return "Fail"

# Run the function with the example
student_marks = [60, 45, 70, 55]
result = process_result(student_marks)
print("Average:", sum(student_marks)/len(student_marks))
print("Result:", result)
```

- Create a cab fare calculator:

  - Peak hour → 25% extra


  - Allow retry using while loop


**Real-Time Use**
- Ride-hailing apps


- Transport systems


**Hint**
- Function for fare calculation


- While loop for retry


```python
def calculate_fare(distance, peak_hour=False):
    base_fare = 50
    per_km_rate = 12

    # Calculate normal fare
    fare = base_fare + (distance * per_km_rate)

    # Apply peak hour surcharge
    if peak_hour:
        fare *= 1.25  # 25% extra

    return fare


def cab_fare_system():
    while True:
        try:
            # Get distance input
            distance = float(input("Enter distance travelled (in km): "))

            # Ask if peak hour
            peak = input("Is it peak hour? (yes/no): ").strip().lower()
            peak_hour = (peak == "yes")

            # Calculate fare
            fare = calculate_fare(distance, peak_hour)
            print(f"Estimated Fare: ₹{fare:.2f}")
            break  # Exit loop after successful calculation

        except ValueError:
            print("Invalid input. Please enter numeric values for distance.")
            # Retry continues automatically due to while loop

# Run the system
cab_fare_system()
```

## **Problem 5: Employee Attendance Eligibility Checker**
**Problem Statement**
- Create a function that:
  - Accepts attendance list ("P", "A")


  - Calculates attendance percentage


  - Returns Eligible if ≥ 75% else Not Eligible


**Real-Time Use**
- HR payroll systems


- Internship tracking


**Hint**
- Use loop to count present days


- Use return statement


```python
def employee_attendance_checker(attendance):
  if not attendance:
    return "No Attendance Provided."

  # Calculate the Present days
  present_days = attendance.count("P")
  print("Present days:", present_days)

  # Calculate Attendance Percentage
  percentage = (present_days / len(attendance)) * 100
  print(f"Attendance Percentage: {percentage:.2f}%")

  # Return Eligible results
  if percentage >= 75:
    return "Eligible."
  else:
    return "Not Eligible."

# Run Function Example
attendance = ["P", "A", "P", "P", "A", "P", "P", "P", "P", "P"]
print("Eligibility Result:",employee_attendance_checker(attendance))
```
